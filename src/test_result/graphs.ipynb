# Error Bar

```python
def load_episode_totals(results_dir_pkl):
    data = train.load_data(results_dir_pkl)

    ep_ph = np.array(data["episode_phonemes"])  # shape: (5000, 3)
    total_per_episode = ep_ph.sum(axis=1)       # shape: (5000,)

    return total_per_episode
```

```python
base_path = "/home/ofir_machlof/env/src/test_result"
baseline_totals_path = f"{base_path}/train_result_2025_11_18_MAX_STEPS=500_MAX_EPISODE=5000_energy_decay/train_20251119.pkl"
# baseline_totals_path = f"{base_path}/train_result_2025_11_18_MAX_STEPS=500_MAX_EPISODE=5000_min_energy_to_speak/train_20251119.pkl"
equality_totals_path = f"{base_path}/train_result_2025_12_03_MAX_STEPS=500_MAX_EPISODE=5000_decay_efficiency/train_20251203.pkl"
```

```python
baseline_totals = load_episode_totals(baseline_totals_path)
equality_totals = load_episode_totals(equality_totals_path)
```

```python
baseline_mean = baseline_totals.mean()
baseline_std  = baseline_totals.std(ddof=1)

equality_mean = equality_totals.mean()
equality_std  = equality_totals.std(ddof=1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ["Baseline", "Equality"]
means = [baseline_mean, equality_mean]
errors = [baseline_std, equality_std]

x = np.arange(len(labels))

plt.figure(figsize=(6,4))
plt.bar(x, means, yerr=errors, capsize=6)
plt.xticks(x, labels)
plt.ylabel("Total phonemes per episode")
plt.title("Speech activity (mean Â± std across episodes)")
plt.tight_layout()
plt.show()
```

```python
last_n = 1000

baseline_tail = baseline_totals[-last_n:]
equality_tail = equality_totals[-last_n:]
```

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ["Baseline", "Equality"]
means = [baseline_tail.mean(), equality_tail.mean()]
errors = [baseline_tail.std(ddof=1), equality_tail.std(ddof=1)]

x = np.arange(len(labels))

plt.figure(figsize=(6,4))
plt.bar(x, means, yerr=errors, capsize=6)
plt.xticks(x, labels)
plt.ylabel("Total phonemes per episode")
plt.title("Speech activity (mean Â± std across episodes)")
plt.tight_layout()
plt.show()
```
